```python
class Trapezoidal:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def mu(self, x):
        a = self.a
        b = self.b
        c = self.c
        d = self.d
        if a<=x<b:
            return (x-a)/(b-a)
        elif b<=x<c:
            return 1
        elif c<=x<d:
            return (d-x)/(d-c)
        return 0
    def alphaCut(self, alpha):
        if alpha<=0:
            return intNum(self.a, self.d)
        elif alpha==1:
            return intNum(self.b, self.c)
        lowLim = self.a + alpha*(self.b-self.a)
        upLim = self.d - alpha*(self.d-self.c)
        return intNum(lowLim, upLim)

lb = Trapezoidal(12, 13, 14, 15)
ub = Trapezoidal(9, 10, 11, 12)
N = 1001
h = 1/N
lowLim = 8
upLim = 16
lf = FuzzySet([FuzzyElement(lowLim + i*h*(upLim- lowLim), lb.mu(lowLim + i*h*(upLim- lowLim))) for i in range(N+1)])
uf = FuzzySet([FuzzyElement(lowLim + i*h*(upLim- lowLim), ub.mu(lowLim + i*h*(upLim- lowLim))) for i in range(N+1)])
f1 = lf.plot()
f2 = uf.plot()
plt.show()
```
